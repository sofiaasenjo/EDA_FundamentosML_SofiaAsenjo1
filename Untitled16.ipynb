```python
%matplotlib inline

from scipy import stats # Tiene un nivel de precision mas alto
import seaborn as sns

np.random.seed(2016) # replicar random

# parametros esteticos de seaborn
sns.set_palette("deep", desat=.6)
sns.set_context(rc={"figure.figsize": (8, 4)})
```

```python
# esto calcula la media de las edad para hacer el histograma
df.mean(numeric_only=True)
df['Age'].mean()
```

```python
#Esto calcula la desviación estandas de la edad
df.std(numeric_only=True)
df['Age'].std()
```

```python
#2
# Medidas de tendencia central
media = df.mean(numeric_only=True)  # Media
mediana = df.median(numeric_only=True)  # Mediana
moda = df.mode(numeric_only=True)  # Moda

# Medidas de dispersión
desviacion_estandar = df.std(numeric_only=True)  # Desviación estándar
varianza = df.var(numeric_only=True)  # Varianza
rango = df.max(numeric_only=True) - df.min(numeric_only=True)  # Rango
iqr = df.quantile(0.75, numeric_only=True) - df.quantile(0.25, numeric_only=True)  # Rango intercuartílico (IQR)

# Mostrar los resultados
print("Medidas de tendencia central:")
print(media)
print(mediana)
print(moda)

print("\nMedidas de dispersión:")
print(desviacion_estandar)
print(varianza)
print(rango)
print(iqr)
```

```python
#3
# Para cada columna numérica en el DataFrame
for columna in df.select_dtypes(include=np.number):
    # 1. Histograma para visualizar la distribución
    plt.figure()  # Crea una nueva figura para cada histograma
    sns.histplot(df[columna])  # Crea el histograma con seaborn
    plt.title(f'Distribución de {columna}')  # Agrega un título
    plt.show()  # Muestra el histograma

    # 2. Calcular la asimetría (skewness)
    skewness = stats.skew(df[columna])
    print(f'Asimetría de {columna}: {skewness}')

    # 3. Interpretar la asimetría
    if skewness > 0:
        print(f'La distribución de {columna} está sesgada hacia la derecha (positiva).')
    elif skewness < 0:
        print(f'La distribución de {columna} está sesgada hacia la izquierda (negativa).')
    else:
        print(f'La distribución de {columna} es aproximadamente simétrica.')
```

```python
#4
# Tabla de frecuencia
bins = [18, 25, 35, 45, 55, 65]  # Define los rangos de edad (ajusta según tus datos)
labels = ['18-24', '25-34', '35-44', '45-54', '55-64']  # Define las etiquetas para los rangos

df['RangoEdad'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
tabla_frecuencia = df['RangoEdad'].value_counts().sort_index()
print(tabla_frecuencia)

# Tabla de frecuencia sin agrupar en rangos
tabla_frecuencia_sin_rangos = df['Age'].value_counts().sort_index()
print(tabla_frecuencia_sin_rangos)
```

```python
#5
# Graficando histograma
mu, sigma = 36.9, 9.1 # media y desvio estandar
datos = np.random.normal(mu, sigma, 1000) #creando muestra de datos

# histograma de distribución normal.
cuenta, cajas, ignorar = plt.hist(datos, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma')
plt.show()
```
